# WordCloud

R&Dウィークリーニュースでword cloud


```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import requests
import MeCab as mc
from os import path
```

```python
def mecab_analysis(text):
#    t = mc.Tagger('-Ochasen -d /usr/local/Cellar/mecab/0.996/lib/mecab/dic/mecab-ipadic-neologd/')
    t = mc.Tagger('-Ochasen')
    enc_text = text
    node = t.parseToNode(enc_text) 
    output = []
    while(node):
        if node.surface != "":  # ヘッダとフッタを除外
            word_type = node.feature.split(",")[0]
            if word_type in ["形容詞", "動詞","名詞", "副詞"]:
                output.append(node.surface)
        node = node.next
        if node is None:
            break
    return output
```

```python
def get_text():
    filename = "../text/test.txt"
    text = open(filename, "r",encoding="utf-8").read()
    text = text.replace('\n','') #改行を削除
    text = text.replace('\r','') #スペースを削除
    
    return mecab_analysis(text)
```

```python
def create_wordcloud(text):

    # 環境に合わせてフォントのパスを指定
    fpath = "C:/Windows/Fonts/YuGothB.ttc"

    # ストップワードの設定
    stop_words = [ u'てる', u'いる', u'なる', u'れる', u'する', u'ある', u'こと', u'これ', u'さん', u'して', \
             u'くれる', u'やる', u'くださる', u'そう', u'せる', u'した',  u'思う',  \
             u'それ', u'ここ', u'ちゃん', u'くん', u'', u'て',u'に',u'を',u'は',u'の', u'が', u'と', u'た', u'し', u'で', \
             u'ない', u'も', u'な', u'い', u'か', u'ので', u'できる', u'よう', u'']

    wordcloud = WordCloud(background_color="white",
                          font_path=fpath, 
                          width=900, 
                          height=500, 
                          stopwords=set(stop_words)
                         ).generate(text)

    plt.figure(figsize=(15,12))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
wordlist = get_text()
create_wordcloud(" ".join(wordlist))
```
