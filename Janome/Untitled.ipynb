# Janomeの基本操作
JanomeはPythonの形態素解析エンジン。

日本語のテキストを形態素ごとに分割して品詞を判定したり分かち書き（単語に分割）したりできる。


## インストール
pip でインストール可能。依存ライブﾗﾘ等なく手軽。

$ pip instrall janome

## データ読み込み
今回はウィークリーニュースのテキストをデータとして利用する。

```python
# ファイルの読み込み
filename = "../text/test.txt"
text = open(filename, "r",encoding="utf-8").read()

print(text)
```

```python
from janome.tokenizer import Tokenizer

t = Tokenizer()

s = 'すもももももももものうち'

print(type(t.tokenize(s)))
# <class 'list'>

print(type(t.tokenize(s)[0]))
```
